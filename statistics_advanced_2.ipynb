 1: What is hypothesis testing in statistics?
  # Hypothesis testing is a statistical method used to determine whether there is enough evidence in a sample data to infer that a certain condition or assumption (the hypothesis) is true for the entire population. It's a formal procedure to decide between two competing hypotheses:
      1.  Null Hypothesis (H₀): This is the default assumption, often stating there is no effect or no difference.
      
      2.Alternative Hypothesis (H₁ or Ha): This is what you are trying to prove, suggesting there is an effect or a difference.


  2: What is the null hypothesis, and how does it differ from the alternative hypothesis?
  # The null hypothesis (H₀) and the alternative hypothesis (H₁) are two competing statements about a population that are tested against each other in hypothesis testing. Here's how they differ:
  
   Null Hypothesis (H₀):
    
    # Definition: It is the default assumption or the status quo. It typically states that there is no effect, no difference, or no relationship between variables in the population.
    # Purpose: It serves as a baseline for comparison. We assume the null hypothesis is true until there is sufficient evidence to reject it.
    # Symbol: Represented as H₀.
    # Examples
      #The average height of men and women is the same.
      #The new drug has no effect on blood pressure.
      #There is no correlation between hours of study and exam scores.

Alternative Hypothesis (H₁ or Ha):

   #Definition: It is the statement that you are trying to find evidence for. It proposes that there is an effect, a difference, or a relationship between variables in the population. It is the opposite of the null hypothesis.
   # Purpose: It represents the researcher's claim or the effect they hope to observe.
   #Symbol: Represented as H₁ or Ha.
   # Examples
Key Differences:

  # Starting Point: We start by assuming the null hypothesis is true and look for evidence to reject it. We do not assume the alternative hypothesis is true from the beginning.
  # Goal: The goal of hypothesis testing is to determine whether there is enough evidence to reject the null hypothesis in favor of the alternative hypothesis.
  # Decision: The decision in hypothesis testing is either to "reject the null hypothesis" or "fail to reject the null hypothesis." We do not "accept the alternative hypothesis" directly; we conclude that there is enough evidence to support it by rejecting the null.
  # Nature of Statement: The null hypothesis often represents a statement of no change or no effect, while the alternative hypothesis represents a statement of change, difference, or effect.
In essence, hypothesis testing is a process of trying to disprove the null hypothesis to provide support for the alternative hypothesis.


 3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.
   # The significance level, often denoted by the Greek letter alpha (α), is a crucial concept in hypothesis testing. It represents the probability of making a Type I error, which is the error of rejecting the null hypothesis when it is actually true.
   Here's a breakdown of its role:

    # Threshold for Decision: The significance level acts as a threshold for deciding whether to reject the null hypothesis.
    # Defining the Rejection Region: It helps define the rejection region in the distribution of the test statistic. If the calculated test statistic falls within this rejection region, we reject the null hypothesis.
    # Balancing Type I and Type II Errors: Choosing a significance level involves a trade-off between Type I errors (rejecting a true null hypothesis) and Type II errors (failing to reject a false null hypothesis). A lower significance level reduces the risk of a Type I error but increases the risk of a Type II error.
    #Interpreting the P-value: The significance level is compared to the p-value (the probability of observing the data if the null hypothesis is true).
      # If the p-value is less than or equal to the significance level (p ≤ α), we reject the null hypothesis. This means that the observed data is unlikely to have occurred by random chance if the null hypothesis were true.
      # If the p-value is greater than the significance level (p > α), we fail to reject the null hypothesis. This means that the observed data is reasonably likely to have occurred by random chance if the null hypothesis were true.
Commonly used significance levels are 0.05 (or 5%) and 0.01 (or 1%). A significance level of 0.05 means that there is a 5% chance of incorrectly rejecting the null hypothesis when it is true.

In summary, the significance level is a pre-determined probability that helps researchers decide how much evidence is needed to reject the null hypothesis and conclude that there is a statistically significant result. It sets the standard for how willing you are to accept the risk of a Type I error.


 4: What are Type I and Type II errors? Give examples of each.
  #In hypothesis testing, there are two main types of errors we can make:
 Type I Error (False Positive):

   # Definition: This occurs when you reject the null hypothesis (H₀) when it is actually true. You conclude there is a significant effect or difference when there isn't one in the population.
   # Probability: The probability of making a Type I error is equal to the significance level (α).
   # Analogy: In a medical test, this is like a test indicating a person has a disease when they actually don't.
   # Example: A pharmaceutical company tests a new drug and concludes it lowers blood pressure (rejecting H₀: the drug has no effect) when, in reality, it has no effect on blood pressure.
Type II Error (False Negative):
   # Definition: This occurs when you fail to reject the null hypothesis (H₀) when it is actually false. You conclude there is no significant effect or difference when there actually is one in the population.
   # Probability: The probability of making a Type II error is denoted by beta (β). The power of a test (1 - β) is the probability of correctly rejecting a false null hypothesis.
   # Analogy: In a medical test, this is like a test indicating a person is healthy when they actually have a disease.
   # Example: A researcher is testing a new teaching method and concludes it doesn't improve student performance (failing to reject H₀: the method has no effect) when, in reality, the method significantly improves performance.

 #The Z-test and T-test are both statistical tests used to compare means, but they are applied in different situations depending on what information you have about the population and the sample size.

 Here's the key difference and when to use each:

 Z-test:
   # When to Use:

     # When the population standard deviation (σ) is known.
     # When the sample size (n) is large (generally n ≥ 30), even if the population standard deviation is unknown (due to the Central Limit Theorem, the sample standard deviation can be a good estimate of the population standard deviation in large samples).
   # Assumption: The data should be normally distributed, or the sample size should be large enough for the Central Limit Theorem to apply.

   # Formula: The test statistic is calculated as:
     Z = (x̄ - μ) / (σ / √n)
    Where
     # x̄ is the sample mean
     # μ is the population mean (under the null hypothesis)
     # σ is the population standard deviation
     #n is the sample size
   Distribution: The Z-test statistic follows a standard normal distribution (Z-distribution).
